```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d-%m-%Y")
```

```python
df[['Amount','Profit','Quantity']].describe()
```

```python
df['Category'].unique()
df['Sub-Category'].unique()
df['PaymentMode'].unique()
```

```python
# create separate Year and Month columns (optional but useful)
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
# remove leading/trailing spaces in text columns
text_cols = ['Order ID','Category','Sub-Category','PaymentMode',
             'CustomerName','State','City','Year-Month']
for col in text_cols:
    df[col] = df[col].str.strip()

# drop exact duplicate rows (if any)
df = df.drop_duplicates()

# ensure numeric columns are correct type
df[['Amount','Profit','Quantity']] = df[['Amount','Profit','Quantity']].astype('int64')
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
